```python
import pandas as pd
import numpy as np
# reading the CSV file
csvFile = pd.read_csv('data.csv')
# displaying the contents of the CSV file
df = pd.DataFrame(csvFile)
#df.drop(columns=['D', 'H'])
#data_top = df.head() 
# print(df)
df.loc[df['H'] !='insn per cycle','G'] = '0'
df
```

```python
li=[]
for i in df['B']:
    li.append(i)
print(len(li))
arr=np.array(li)
arr.resize(391,21)
arr.shape
arr
```

```python
df_new = pd.DataFrame(arr, columns =['node-load-misses','node-store-misses','branch-load-misses','branch-misses','iTLB-load-misses','dTLB-load-misses','dTLB-store-misses','L1-icache-load-misses','L1-dcache-load-misses','l2_rqsts.code_rd_miss','l2_rqsts.all_demand_miss','l2_rqsts.rfo_miss','cache-misses','LLC-load-misses','LLC-store-misses','instructions','cycles','L1-dcache-loads','L1-dcache-load-misses1','LLC-loads','LLC-load-misses1'])
df_new = df_new.replace('<not counted>','0')  
del df_new['L1-dcache-load-misses1']
del df_new['LLC-load-misses1']
del df_new['L1-dcache-loads']
del df_new['LLC-loads']
del df_new['instructions']
del df_new['cycles']
del df_new['node-load-misses']
del df_new['node-store-misses']
df_new
```

```python
li2=[]
for i in df['G']:
    li2.append(i)
li3=[]
for i in range(15,len(li2),21):
    li3.append(li2[i])
df_new['CPI']=li3
df_new
```

```python
# print(df_new['CPI'].astype(float).mean())
for i in list(df_new):
    df_new[i]=df_new[i].replace('0',df_new[i].astype(float).mean())

df_new
```
